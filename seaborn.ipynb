```python
import tensorflow as tf
import numpy as np
normal_logins = np.array([3,4,3,5,4,4,3,5,4,100,2000], dtype=float)
threshold = tf.constant(10.0)
for value in normal_logins:
    if value > threshold:
        print(value, "ALERT: Suspicious Login Activity!")
    else:
        print(value, "Normal")
```

```python
logs = ["login failed", "login failed", "login failed",
        "login failed", "login failed", "login success"]

fail_count = 0

for log in logs:
    if log == "login failed":
        fail_count += 1
        if fail_count >= 5:
            print("Intrusion Detected! Too many failed logins!")
            break
    else:
        print("No issue")
```
